# Histograms

Let's run the following code in order to see a basic histogram in R.

```
x = c(1,1,2,3)
hist(x, breaks=seq(from=0.5, to=3.5, by=1), freq=FALSE, cex.axis=2)
```

For supplying this argument, it is potentially useful to know that the `seq` function in R can be used to generate a *seq*uence of evenly spaced values.

`seq(from=a, to=b, by=c)` will yield a vector starting at `a`, ending by `b`, and spaced apart by increment `c`.

```
seq(from=0, to=1, by=0.25)
```

`seq(from=a, to=b, length.out=d)` will yield a vector starting at `a`, ending at `b`, and with exactly `d` elements in the vector.

```
seq(from=0, to=1, length.out=6)
```
