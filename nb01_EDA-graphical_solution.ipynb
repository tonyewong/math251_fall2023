# Boxplots

We can use the `boxplot()` function to create a boxplot of the Old Faithful eruption times:

```
boxplot(faithful[,1], ylab="Eruption time (minutes)")
```

### Task 6

Using the boxplot, estimate the following statistics of the eruption data:
* median
* first and third quartiles

Can you determine the minimum and maximum values? Why or why not?

**SOLUTION:**

Median:  given by the solid line in the middle of the box, the median is about 4 minutes

Q1 is the bottom of the box, at about 2.2 minutes

Q3 is the top of the box, at about 4.5 minutes

The whiskers extend to the farthest data points 1.5$\times$ the IQR, and fliers are depicted as dots or X's outside of the whiskers. Since there aren't any fliers, we know all of the data must be within the whiskers. So, the extent of the whiskers gives the min and max values.

Min is given by the lower whisker, at about 1.6 minutes.

Max is given by the upper whisker, at about 5.1 minutes.

### Task 7

Based **only** on the boxplot above, do you think the eruption time data are left-skewed or right-skewed? Why? (Make an argument based on the features of the boxplot)

**SOLUTION:**

The median is closer to Q3 than to Q1. This suggests that the data are "piled up" towards higher values, since there are an equal number of data points in the small area between the median and Q3 (4-4.5 mins) as there are in the larger area between Q1 and the median (2.2-4 mins). This corresponds to left-skewed data.
