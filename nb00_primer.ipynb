## 2  Code cells and Markdown cells

Jupyter notebooks are made up of different "cells". Click once on this section of text. A box will appear around this text (and some areas above/below it) - all of that is within this cell. In the upper-right corner of the cell, you will see a horizontal menu with options to do various things with this cell, including move it up/down, edit it, add comments to it or delete it (among other things). Hover your cursor over each of the options to see what magic you can perform. Note that adding comments to the cell works analogously to adding comments to a Google Doc.

There are two types of cells:  **Markdown** cells and **code** cells. Markdown is a kind of text formatting language that makes things look pretty when viewed in a web browser. You can also include adornments like hyperlinks, LaTeX equations, section headers and images.

What you are looking at right now is the nicely-formatted Markdown text. To see the raw Markdown, do one of:
* click on the `Edit` option in the menu of the cell you have selected,
* press `Enter` while that cell is selected, or
* double-click on that cell.

A couple of the features you will see in this Markdown cell:
* The `<br>` command adds a vertical space to your notebook. This is nice for separating sections.
* The `---` command creates a horizontal line. This is also nice for separating sections.
* The `#` sign is used to create a new section header; two `#` signs (`##`) is used to create a new subsection header; `###` creates a subsubsection; more than that and you're just showing off.
* You can create a bulleted list by using the asterisk `*` and a space.
* You can create regular text by just typing as usual.
* You can create bold-faced text by wrapping it with two asterisks on either side (see the words "Markdown" and "code" above).
* To create a new paragraph, you must include a blank line between the old and new paragraphs.

At this point you might be wondering how to turn this cell back into the fully formatted Markdown text instead of the raw Markdown you're probably still looking at. In Colab, you have two options:
1. Press `Shift+Enter` to *execute* the cell.
1. Click the `Close Markdown editor` button in the cell menu in the upper-right.

## 3  Basic mathematical operations

Now arguably the most important part of the whole computer coding part of this course is to perform mathematical operations. It begins by defining **variables**, which work the same way as defining a variable in regular paper-and-pencil mathematics. In R, the "arrow" `<-` has traditionally taken the place of an equals sign to assign a value to a variable. But for our purposes, a regular equals sign will work fine. So, for example, if we would like to declare that the variable `x` is equal to 6, we can say:

```
x = 6
```

To actually enter that into the computer, you need to use the second kind of Jupyter notebook cell, a **code cell**. You can place one just below your currently selected cell by clicking `+ Code` in the upper-left, or by hovering your cursor at the bottom of this cell and then clicking `+ Code`. For convenience, a code cell is included below with the code to set a variable named `x` equal to 6. To *execute* that code cell, select the cell by clicking on it; the thin box should surround the cell. Then, hit `Shift+Enter` or push the "Play" button on the left.

````
x = 6
````

After you have executed the above cell, we can check to see what value the variable `x` has in the computer's memory by asking R to print `x` to the screen. We use a **function** called (appropriately enough) `print()` to do this. Inside the parentheses, we place the **argument** of the function. In mathy terms, we call this the **input** to the function. In this case, the argument (input) should be `x`, since that's the thing we want to be printed. Execute the code cell below to run the `print()` function for the input variable `x`.

````
print(x)
````

Armed with our knowledge of how to define a variable, we can do all kinds of mathematical operations to these variables. For example, to add 4 to `x`, we can do exactly what you would think you should do:

````
x + 4
````

We can define a new variable, `y`, whose value is the value of `x` plus 4 as follows. Note that you can execute multiple commands within a single code cell by putting a new command on a new line (using the `Enter` key to create new lines). For example:

````
y = x + 4    # create a variable called y and assign it the value x+4
print(y)     # print y to the screen
````

**Sanity Check:**  I love sanity checks. There's a great way to make sure your answers are not insane.

* $2^3 = 8$
* $3^3 = 27$

## 4 A little bit fancier mathematical operations

Suppose we measure the maximum monthly temperature (in degrees Fahrenheit) in a city throughout a year and obtain the following _array_ of data:

```
[39, 45, 54, 66, 72, 89, 94, 90, 81, 79, 57, 41]
```

That is, the maximum temperature in January was 39 $^\circ$F, the maximum temperature in February was 45 $^\circ$F, and so on. How can we represent this set of data in R?

Well, for me, when I Googled "array in R" one of the first results was this:

https://www.tutorialspoint.com/r/r_arrays.htm

That tutorial points to the simplest way to create an array (or a set of related variables) in R, which is with the `c()` function. The argument of this function is a list of the values you want in the array, separated by commas. To set a variable called `temps` to this array, we can do the following:


````
temps = c(39, 45, 54, 66, 72, 89, 94, 90, 81, 79, 57, 41)
````

Let's print `temps` to the screen to check that it has the right numbers, and all of the numbers, stored in it.

````
print(temps)
````

Great! It looks like things are working well.

Are you wondering why the heck the function to create an array out of a set of numbers is called "c"? Why the letter c? This is a reference to the fact that the `c()` function in R is a generic function to _combine_ its arguments, or _concatenate_ them. How cool, right? Tell your friends!

<br>

Now, how do we access each element of the array individually? If someone asks us what the maximum temperature in January was, how do we obtain that from our variable `temps`?

We can access the $i$th value in `temps` by **indexing** within the array. That is, in math-speak, $\texttt{temps}_i$ (with the subscript $i$) typically refers to the $i$th element within the array `temps`. In R, we grab the $i$th element by following the array name with square brackets:  `temps[i]`. Since January is the first, or $i=1$, element within `temps`, we do `temps[1]`:

````
print(temps[1])
````

We can also do things like create a plot of the temperatures.

````
plot(temps)
````

That's nice and all, but it lacks a critical aspect of a nice plot: axis labels! We can use the `plot()` function's *optional arguments* for `xlab` and `ylab` to provide names for the x-axis label and y-axis label, respectively. These arguments are a type of variable called a **string**, which is a fancy way to say that they're words (or more specifically, a set of characters _strung_ together). We denote string variables by wrapping them in quotation marks "_string characters here_")

````
plot(temps, xlab="month", ylab="degrees Fahrenheit")
````

Ah. Much better. There are loads of other ways you can customize a plot in R. For example, to connect these dots and plot the temperatures as a _line_, we would use the `type` argument of `plot()`:

````
plot(temps, xlab="month", ylab="degrees Fahrenheit", type="l")
````

The `type` argument can have a variety of different values. Here, we are using `type="l"` to create a line plot (the "`l`" stands for "line").

## 5 Formatting mathematical expresssions
