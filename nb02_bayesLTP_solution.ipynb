````
# SOLUTION:

num_samples = 10000  # how many samples do we want to use?
num_R = 0         # initialize a counter for how many times Artemis picks Red
num_R_B1 = 0      # initialize a counter for how many times Artemis picks Red AND from Box 1

for (i in 1:num_samples) {
  box = sample(c(1,2), size=1)   # a single trial of picking a box totally at random
  if (box==1) {
    color = sample(c("R","G"), prob=c(7/9,2/9), size=1)  # given picking from Box 1, what color does she pick?
  } else if (box==2) {
    color = sample(c("R","G"), prob=c(3/7,4/7), size=1)  # given picking from Box 2, what color does she pick?
  }
  if (color=="R") {
    num_R = num_R+1      # if ball is Red, add 1 to the count for how many Reds
  }
  if (color=="R" && box==1) {
    num_R_B1 = num_R_B1+1   # if ball is Red and picking from Box 1, add 1 to that count
  }
}
Pr_R_and_B1 = num_R_B1/num_samples    # estimate P(Red and Box 1)
Pr_R = num_R/num_samples           # estimate P(Red)
Pr_B1_given_R = Pr_R_and_B1/Pr_R  # estimate P(Box 1 | R)
print(Pr_B1_given_R)
````
